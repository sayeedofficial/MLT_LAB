```python
#EM Algorithm
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
iris = load_iris()

x = pd.DataFrame(iris.data)
x.columns = ['Sepal_Length','Sepal_Width','Petal_Length','Petal_Width']

y = pd.DataFrame(iris.target)
y.columns = ['Targets']
```

```python
kmns = KMeans(n_clusters=3)
kmns.fit(x)
y_kmeans = kmns.fit_predict(x)
```

```python
gmm = GaussianMixture(n_components=3,random_state=0)
gmm.fit(x)
y_gmm = gmm.predict(x)
```

```python
print('The accuracy score of K-Mean: ',accuracy_score(y, kmns.labels_))
print('The Confusion matrixof K-Mean: \n',confusion_matrix(y, kmns.labels_))
```

```python
print("Accuracy Score ",accuracy_score(y,y_gmm))
print("Confusion Matrix ",confusion_matrix(y,y_gmm))
```

```python
plt.figure(figsize=(30,7))
colormap = np.array(["red","Lime",'black'])
```

```python

# Plot the Original Classifications
plt.subplot(1,3,1)
plt.scatter(x.Petal_Length, x.Petal_Width, c=colormap[y.Targets], s=40)
plt.title('Real Classification')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')

# Plot the K Means Classifications
plt.subplot(1,3,2)
plt.scatter(x.Petal_Length, x.Petal_Width, c=colormap[kmns.labels_], s=40)
plt.title('K Mean Classification')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')

# Plot the K Means Classifications
plt.subplot(1,3,3)
plt.scatter(x.Petal_Length, x.Petal_Width, c=colormap[y_gmm], s=40)
plt.title('GMM Classification')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
```
